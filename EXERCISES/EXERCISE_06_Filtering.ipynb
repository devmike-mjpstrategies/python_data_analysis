# Filtering Data Exercise
Answer the following questions using the `bestsellers.csv` dataset

```python
import pandas as pd
besties = pd.read_csv("../DATA/bestsellers.csv")
besties.info()
```

```python
#Find the books written by Pete Souza
besties[besties["Author"] == "Pete Souza"]
```

```python
# Find the books that are under 10 dollars
besties[besties["Price"] < 10]
```

```python
#Find the books that have a price between 50 and 60 dollars
besties[besties["Price"].between(50, 60)]
```

```python
#Find all the books written by Kristin Hannah, Andy Weir, or Delia Owens
# kristin = besties["Author"] == ("Kristin Hannah")
# andy = besties["Author"] == ("Andy Weir")
# delia = besties["Author"] == ("Delia Owens")
# besties[kristin | andy | delia]

besties[besties["Author"].isin(["Kristin Hannah", "Andy Weir", "Delia Owens"])]
```

```python
#Find the Non Fiction books that are rated 4.9
nonfic = besties["Genre"] == "Non Fiction"
high_rating = besties["User Rating"] == 4.9
besties[nonfic & high_rating]
```

```python
#Find the fiction book with the lowest User Rating
besties[besties["Genre"] == "Fiction"].sort_values("User Rating")

```

```python
#Find 2012's top 5 Fiction books with the most Reviews
besties[(besties["Genre"] == "Fiction") & (besties["Year"] == 2012)].sort_values("Reviews", ascending=False).head()
```

```python
#Create a bar plot showing the 5 authors who have the most books with a rating under 4.5
# My solution 
low_rated = besties[besties["User Rating"] < 4.5]
low_rated.value_counts("Author").head().plot(kind="bar")

# Instructor solution
# low_rated = besties[besties["User Rating"] < 4.5]
# low_rated["Author"].value_counts().head().plot(kind="bar")
```
