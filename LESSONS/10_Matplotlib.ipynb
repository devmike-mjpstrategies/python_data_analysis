## Bar Plots

```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10,25,5,30,21]
germinated = [74, 88, 56,69,59]
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, germinated, color="orchid")
plt.title("Germination Rates")
plt.ylabel("Number Of Seeldings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="red")
plt.title("Failure Rates")
plt.ylabel("Number Of Dead Seedlings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db")
plt.bar(plants, germinated, width=0.4, color="#2ecc71", align="edge")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db", bottom=germinated )
plt.bar(plants, germinated, color="#2ecc71")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db", bottom=germinated,label="Failed Seedlings")
plt.bar(plants, germinated, color="#2ecc71", label="Germinated Seedlings")
plt.legend(shadow=True, frameon=True, facecolor="white")
plt.title("Greenhouse Seedlings")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.barh(plants, germinated, color="#2ecc71")
```

```python
import pandas as pd
titanic = pd.read_csv("../DATA/titanic.csv")
titanic["age"] = pd.to_numeric(titanic["age"], errors="coerce")
```

```python
plt.hist(titanic["age"])
```

```python
first_class = titanic[titanic["pclass"] == 1]["age"]
second_class = titanic[titanic["pclass"] == 2]["age"]
third_class = titanic[titanic["pclass"] == 3]["age"]
plt.hist(first_class, label="1st Class", alpha=0.5, color="#f1c40f", bins=20)
plt.hist(second_class, label="2nd Class", alpha=0.5, color="#3498db", bins=20)
plt.hist(third_class, label="3rd Class", alpha=0.5, color="#e74c3c", bins=20)
plt.legend()
plt.title("Titanic Passenger Ages")
plt.xlabel("Passenger Age")
```

## Subplots

```python
nums = np.arange(5)
plt.figure(figsize=(10,4))
plt.suptitle("Our First Subplot", fontsize=30)

plt.subplot(1,3,1)
plt.title("X")
plt.plot(nums, nums)

plt.subplot(1,3,2)
plt.plot(nums, nums**2, color="red")
plt.title("X Squared")

plt.subplot(1,3,3)
plt.plot(nums, nums**3, color="yellow")
plt.title("X Cubed")
plt.tight_layout()
plt.show()
```

```python
nums = np.arange(5)
plt.figure(figsize=(4,10))
plt.suptitle("Our First Subplot", fontsize=30)

plt.subplot(3,1,1)
plt.title("X")
plt.plot(nums, nums)

plt.subplot(3,1,2)
plt.plot(nums, nums**2, color="red")
plt.title("X Squared")


plt.subplot(3,1,3)
plt.plot(nums, nums**3, color="yellow")
plt.title("X Cubed")
plt.tight_layout()
plt.show()
```

```python
first_class = titanic[titanic["pclass"] == 1]["age"]
second_class = titanic[titanic["pclass"] == 2]["age"]
third_class = titanic[titanic["pclass"] == 3]["age"]

plt.figure(figsize=(10,4))
ax = plt.subplot(1,3,1)
plt.hist(first_class, label="1st Class", color="#f1c40f", bins=20)
plt.title("1st Class")

plt.subplot(1,3,2, sharey=ax)
plt.hist(second_class, label="2nd Class", color="#3498db", bins=20)
plt.title("2nd Class")


plt.subplot(1,3,3, sharey=ax)
plt.hist(third_class, label="3rd Class", color="#e74c3c", bins=20)
plt.title("3rd Class")
```

## The Object-Oriented Approach

```python
fig, axs = plt.subplots(1,2)
```

```python
axs[0].hist(first_class)
axs[1].hist(second_class)
axs[0].hist(third_class)
axs[0].set_title("My Title")
axs[0].set_xlabel("Age")
axs[1].set_title("My Other Title")
fig
```

```python
fig, axs = plt.subplots(2,2)
axs[0][0].plot(nums,color="red")
axs[0][1].plot(nums*nums)
axs[1][0].plot(nums**3)
axs[1][1].plot(1/nums)

axs[0][1].set_xticks([0,2,4])
```
